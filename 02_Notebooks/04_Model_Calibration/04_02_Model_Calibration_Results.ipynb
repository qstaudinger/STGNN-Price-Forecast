# 1. Load Libraries 

```python
import json
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Results E5

```python
indir = Path("/home/qusta100/STGNN/04_Results/01_Temporary/02_Trial_Results/04_Model_Calibration/04_01_Model_Calibration_HP/02_E10")

rows = []
for fp in sorted(indir.glob("trial_*.json")):
    with fp.open() as f:
        d = json.load(f)
    row = {
        "file": fp.name,
        "trial": d["trial"],
        "score": d["score"],
    }
    row.update(d["hyperparams"])
    rows.append(row)

E5_Time_5N = pd.DataFrame(rows)
E5_Time_5N = E5_Time_5N.sort_values("score").reset_index(drop=True)

E5_Time_5N.head(10)
```

```python
best_E5_Time_5N = E5_Time_5N.iloc[0]
best_E5_Time_5N
```

```python
# Example of (useless) Graphs
plt.figure(figsize=(6, 4))
E5_Time_5N.boxplot(column="score", by="embed_dim")
plt.xlabel("embed_dim")
plt.ylabel("score")
plt.title("Score by embed_dim")
plt.suptitle("")
plt.show()
```
