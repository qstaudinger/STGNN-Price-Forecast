```python
# Global Import and Directories
import sys
sys.path.append("..")
import Setup
from Setup import *
```

## 2. Results E5

```python
indir = Path(
    Setup.RESULTS_DIR / "01_Temporary" / "02_Trial_Results" / "04_Model_Calibration" / "01_E5"
)

rows = []
for fp in sorted(indir.glob("trial_*.json")):
    with fp.open() as f:
        d = json.load(f)
    row = {
        "file": fp.name,
        "trial": d["trial"],
        "score": d["score"],
    }
    row.update(d["hyperparams"])
    rows.append(row)

E5_Time_5N = pd.DataFrame(rows)
E5_Time_5N = E5_Time_5N.sort_values("score").reset_index(drop=True)

E5_Time_5N.head(10)
```

```python
# Show best configuration
best_E5_Time_5N = E5_Time_5N.iloc[0]
best_E5_Time_5N
```

```python
outdir = Path(
    Setup.RESULTS_DIR
    / "01_Temporary"
    / "04_Configurations"
    / "04_Model_Calibration"
    / "01_E5"
)

outdir.mkdir(parents=True, exist_ok=True)

# best Hyperparameters
best_params = E5_Time_5N.loc[
    0,
    E5_Time_5N.columns.difference(["file", "trial", "score"])
].to_dict()

# Save
outpath = outdir / "best_hyperparams.json"
with outpath.open("w") as f:
    json.dump(best_params, f, indent=2)

print("Saved to:", outpath)
```
